```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Transfer_Budget_Data - Sheet1 (1).csv")

teamname = (data['Team_Name'])
budget = (data['Transfer_Budget'])


plt.bar(teamname, budget, color='blue')



plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Transfer budgets per team')

plt.xticks(rotation=45, ha='right')




plt.show()
```
